# B) Data Visualization:

<b>a. Build a bar-plot for the ’InternetService’ column:
</b>

<b>i. Set x-axis label to ‘Categories of Internet Service’</b>

<b>ii. Set y-axis label to ‘Count of Categories’</b>

<b>iii. Set the title of plot to be ‘Distribution of Internet Service’</b>

<b>iv. Set the color of the bars to be ‘orange’</b>

```python
A = customer_churn['InternetService'].value_counts().keys().tolist()
A
```

```python
B = customer_churn['InternetService'].value_counts().tolist()
B
```

```python
plt.bar(A,B, color = 'orange')
plt.xlabel('Categories of Internet Service')
plt.ylabel('Count of Categories')
plt.title('Distribution of Internet Service')
plt.grid()
plt.show()
```

<b>b. Build a histogram for the ‘tenure’ column</b>

<b>i. Set the number of bins to be 30</b>

<b>ii. Set the color of the bins to be ‘green’</b>

<b>iii. Assign the title ‘Distribution of tenure’</b>

```python
plt.hist(customer_churn['tenure'], bins = 30, color = 'green')
plt.title('Distribution of tenure')
plt.show()
```

<b>c. Build a scatter-plot between ‘MonthlyCharges’ & ‘tenure’. Map ‘MonthlyCharges’ to
the y-axis & ‘tenure’ to the ‘x-axis’:</b>

<b>i. Assign the points a color of ‘brown’</b>

<b>ii. Set the x-axis label to ‘Tenure of customer’</b>

<b>iii. Set the y-axis label to ‘Monthly Charges of customer’</b>

</b>iv. Set the title to ‘Tenure vs Monthly Charges’</b>


```python
X = customer_churn['tenure']
Y = customer_churn['MonthlyCharges']
plt.scatter(X,Y, color = 'brown')
plt.xlabel('Tenure of customer')
plt.ylabel('Monthly Charges of customer')
plt.title('Tenure vs Monthly Charges')
plt.show()
```
